```python
import pandas as pd
```

```python
#Leemos uno de los archivos
#dropeamos la columna de la marca temporal
df1 = pd.read_csv('PV/df1.csv')
df1.drop(['Marca temporal'], axis=1, inplace = True)
shape1 = df1.shape
print('Shape del df1 : ' ,shape1)

#Traemos el segundo archivo
df2 = pd.read_csv('PV/df2.csv')
print('Shape del df2 : ' ,df2.shape)
```

```python
#Nos quedamos solo con los puntos verdes del df1, sacando los ruleros
Rulero_df1_idx=df1[df1["Tipo de punto"]=='Rulero'].index
df1=df1.drop(Rulero_df1_idx)

#Verificamos el shape ahora del df1
shape2 = df1.shape
print('Shape del df1 : ' ,shape2)
print('Se eliminaron : ', shape1[0] - shape2[0], 'ruleros')
```

```python
#Acá tenemos el Data Frame final
df_final = pd.merge(df1, df2, how= 'left', on=['ID'])

#Veamos las dimensiones
print(df_final.shape)
print('Perfecto, la cantidad de entradas es la misma que el df1 y se suman los features del df2')

#Visualizamos el df final
df_final
```

```python
#¿Qué columnas tenemos ahora?
df_final.columns
```

```python
#Podemos eliminar los ultimos features, las coordenadas geográficas y la columna 'Calidad_residuo' dado que resultaba de hacer
#los promedios de los puntos verdes.
df_final.drop(['Y_WGS', 'X_WGS', 'Calidad_residuo'], axis=1, inplace= True)
```

```python
#Chequeamos ahora
df_final.dtypes
```

```python
#Creamos la variable TARGET o variable a predecir, que resulta la suma de clasificaciónNum y CantidadNum
#Antes reemplazamos las comas por los puntos y convertimos los datos a int
df_final['ClasificacionNum'] = df_final['ClasificacionNum'].str.replace(",", ".").astype(float)
df_final['CantidadNum'] = df_final['CantidadNum'].str.replace(",", ".").astype(float)
```

```python
df_final['target'] = df_final['ClasificacionNum'] + df_final['CantidadNum']
df_final
```

```python
#Ya tenemos el target, podemos eliminar las columnas 'ClasificacionNum' y 'CantidadNum'
df_final.drop(['ClasificacionNum','CantidadNum'], axis=1, inplace= True)
```
