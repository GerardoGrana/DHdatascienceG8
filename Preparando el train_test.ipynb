# Preparando train_test

```python
#Importamos librerias base
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
#Levanto los dataframes que voy a utilizar
PVR_observaciones = pd.read_csv('Data/PVR_observaciones.csv', encoding='latin1')
PVR_DAM_geodata = pd.read_csv('Data/PVR_DAM_geodata.csv', delimiter= ';')

#Borro los Nans de observaciones
PVR_observaciones.dropna(how='all', axis=1, inplace=True) 
PVR_observaciones.dropna(how='all', inplace= True)

print('Al comienzo: ', PVR_DAM_geodata['punto_tipo'].value_counts())
print('')
#Extraigo las esquinas quedandome solo con puntos verdes y ruleros en PVR_DAM_geodata
#Eliminamos las intersecciones
DAM_to_predict = PVR_DAM_geodata[PVR_DAM_geodata['punto_tipo'] == 'Intersercción']
PVR_geodata = PVR_DAM_geodata.drop(PVR_DAM_geodata[PVR_DAM_geodata['punto_tipo'] == 'Intersercción'].index)
PVR_observaciones = PVR_observaciones[['ClasificacionNum','ID']]
print('')
print('Finalmente: ',PVR_geodata['punto_tipo'].value_counts())
```

```python
PVR_geodata
```

```python
DAM_to_predict.dtypes
```

```python
#Genero el DataFrame final
PVR_train_test = pd.merge(PVR_observaciones, PVR_geodata , how="left", on=["ID"]) 
PVR_train_test
```

```python
#Miremos los features y los nulos
print(PVR_train_test.isnull().sum())
#Eliminamos los nulls en todas las columnas
PVR_train_test.dropna(axis=0, inplace= True)
```

```python
#Miramos nuevamente los nulos
print(PVR_train_test.isnull().sum())
print('')
print('Se eliminaron correctamente :)')
```

```python
#Binarizamos la columna de calidad 
diccionario = {1.0: 'Apto', 2.0: 'NoApto', 3.0 : np.NaN}
PVR_train_test['ClasificacionNum']= PVR_train_test['ClasificacionNum'].map(diccionario)
PVR_train_test = PVR_train_test.rename(columns={'ClasificacionNum':'Target'})
#Eliminamos los NaN
PVR_train_test.dropna(axis=0, inplace=True)
```

```python
#Vamos a eliminar columnas que son irrelevantes o que tienen mucha correlación con otras
PVR_train_test_copy = PVR_train_test.copy()
PVR_train_test.drop(['ID', 'index','x','y', 'punto_tipo'], axis=1, inplace= True)


#educativos = ['inicial','primario','secundario','egb', 'polimodal', 'superior', 'universitario', 'post_universitario', 'lee_y _esc', 'NoLee_ni_e']
```

### Reclasificación de variable punto_tipo
Dentro del dataframe DAM_to_predict la columna punto_tipo contiene el valor 'Intersección'.
Para realizar la predicción, se reclasificará este valor con dos alternativas: 'Punto verde' y 'Rulero'.
El objetivo de esto es poder analizar la predicción de la intersección preguntándole al algoritmo cómo se comportaría ese lugar si fuera PR o R.
Por lo tanto se duplicará el df DAM_to_predict (copy):
* DAM_to_predict_R
* DAM_to_predict_PV
Luego se apilarán ambos df en uno único llamado DAM_predictions.

```python
DAM_to_predict_R = DAM_to_predict.copy(deep=True)

diccionarioR = {'Intersercción':'Rulero'}
DAM_to_predict_R['punto_tipo'] = DAM_to_predict_R['punto_tipo'].map(diccionarioR)
```
