```python
import pandas as pd
import os
import glob
```

```python
#Traemos el primer archivo
#dropeamos la columna de la marca temporal
df1 = pd.read_csv('Data/df1.csv')
df1.drop(['Marca temporal'], axis=1, inplace = True)
shape1 = df1.shape
print('Shape del df1 : ' ,shape1)

#Traemos el segundo archivo
df2 = pd.read_csv('Data/df2.csv')
print('Shape del df2 : ' ,df2.shape)
```

```python
df2
```

```python
#Acá tenemos el Data Frame final
df_final = pd.merge(df1, df2, how= 'left', on=['ID'])

#Visualizamos el df final
df_final
```

```python
#¿Qué columnas tenemos ahora?
df_final.columns
```

```python
df_final.iloc[5]
```

```python
#Eliminamos algunos features
df_final.drop(['Y_WGS', 'X_WGS', 'Calidad_residuo', '¿Se realiza recolección en el momento?', 'Clasificacion','Cantidad', 'PuntoTipo', 'ID', 'Calidad (grupos)','PuntoNumer', 'PuntoNombr', 'Direccion','link', 'Clasifica','Nombre'], axis=1, inplace= True)
```

```python
#¿Qué columnas tenemos ahora?
df_final.columns
```

```python
#Reemplazamos las comas por los puntos y convertimos los datos a int para los targets
df_final['ClasificacionNum'] = df_final['ClasificacionNum'].str.replace(",", ".").astype(float)
df_final['CantidadNum'] = df_final['CantidadNum'].str.replace(",", ".").astype(float)
df_final['ClasificacionNum'].dtypes
```

```python
#Hacemos un cambio de los valores de ClasificacionNum
eq = {1.0 : 30, 3.0 : 10, 2.0 : 20}
df_final['ClasificacionNum2'] = df_final['ClasificacionNum'].map(eq)
df_final
```

```python
df_final.drop(['ClasificacionNum'], axis=1, inplace= True)
```

```python
df_final
```

```python
df_final.dtypes
```

```python
#Pasamos a float densidad de poblacion
df_final['dens pob'] = df_final['dens pob'].str.replace(",", ".").astype(float)
```

```python
df_final.dtypes
```

```python
#Eliminamos datos nulos
df_final = df_final.dropna()
df_final.isnull().sum()
```

```python
df_final
```

```python
#Creamos la variable target final como la suma de la calidad y la cantidad
df_final['target'] = df_final['CantidadNum'] + df_final['ClasificacionNum2']
```

```python
#Ahora podemos dropear CantidadNum y ClasificacionNum2 y dejar solo el target como variable a predecir
df_final.drop(['CantidadNum', 'ClasificacionNum2'], axis=1, inplace= True)
df_final
```

```python
#Hacemos dummies con las variables que corresponden
df_final2 = pd.get_dummies(df_final, drop_first= True)
df_final2
```

```python
df_final2.shape
```

```python
df_final2.columns
```

```python
df_final2['target'].value_counts() / df_final.shape[0]
```

```python
#Eliminamos las etiquetas que tienen 1 solo valor
df_final2 = df_final2.drop(df_final2[df_final2['target']==23].index)
df_final2 = df_final2.drop(df_final2[df_final2['target']==12].index)
```

```python
df_final2['target'].value_counts() / df_final.shape[0]
```

```python
#Terminamos la limpieza asi que guardamos el df_final2
df_final2.to_csv("df_final2.csv", index=False, encoding='utf-8-sig')
```
